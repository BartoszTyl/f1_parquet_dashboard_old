```python
session = ff1.get_session(2025, "China", "R")
session.load()
# plotting.get_team_color("Red Bull", session)
```

```python
import pandas as pd

# Define the number of laps
number_of_laps = int(session.laps["LapNumber"].max())

# Create an empty list to store the data
laps_data = []

# Loop over each driver and lap
for driver in session.results["Abbreviation"]:
    for lap in range(1, number_of_laps + 1):
        try:
            # Check if the lap has telemetry data for the driver
            lap_data = session.laps.pick_laps(lap).pick_drivers(driver)
            
            if lap_data.empty:
                print(f"No telemetry data for driver {driver} in lap {lap}")
                continue  # Skip this lap and driver if no data is available
            
            # Get the telemetry data for the specific lap and driver
            telemetry_df = pd.DataFrame(lap_data.get_telemetry())
            
            # Add extra columns to keep track of the driver and lap
            telemetry_df['driver'] = driver
            telemetry_df['lap'] = lap
            
            # Append the telemetry data to the list
            laps_data.append(telemetry_df)
        except Exception as e:
            print(f"Error for driver {driver} in lap {lap}: {e}")

# Combine all the data into one large DataFrame
if laps_data:
    all_laps_data = pd.concat(laps_data, ignore_index=True)
    
    # Save the DataFrame as a Parquet file
    all_laps_data.to_parquet('all_laps_telemetry.parquet', index=False)
    # print("Data has been saved as a Parquet file.")
    print("Data has been successfully saved!")
else:
    print("No telemetry data available to save.")
```
